## 10. Столбец с максимальным средним значением после нормировки признаков через стандартное отклонение?
***

Нормируем признак (для числовых признаков) через стандартное отклонение:

```python
import math
numeric = ['age', 'absences', 'G1', 'G2', 'G3']
for column in numeric:
    #print('before:\n', df[column])
    df[column] = (df[column] - df[column].mean())/(math.sqrt(df[column].var()))
    #print('after:\n', df[column])
```

Также преобразуем текстовые значения в числовые:

```python
texted = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']
for column in texted:
    cur_set = set(df[column])
    i = 0
    for item in cur_set:
        df[column] = df[column].replace(item, i)
        i = i + 1
```

```python
df
```

Найдем максимальное среднее значение (среди столбов с числовым значением):

```python
max_average = -math.inf
max_name = ''
for column in numeric:
    if df[column].mean() > max_average:
        max_name = column
        max_average = df[column].mean()
print(f"Максимальное среднее значение {max_average} имеет столбец {max_name}")
```

По результам работы функции, выполненной в блоке выше, видим, что столбец с максимальным средним значением после нормировки признаков - **age**.

## 13. Между какими признаками наблюдается линейная зависимость (корреляция)?
***

Для выявления линейной зависимости признаков (мультиколлениарности), будем использовать VIF (Variable Inflation Factors). 
Теоритические данные взяты [отсюда](https://www.analyticsvidhya.com/blog/2020/03/what-is-multicollinearity/).

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["column_name"] = df_for_training.columns
vif_data["VIF"] = [variance_inflation_factor(df_for_training.values, i) for i in range(len(df_for_training.columns))]
#vif_data = vif_data.transpose()
```

Выведем признаки со значением VIF $\geq 10$:

```python
vif_data.loc[vif_data['VIF'] >= 10.0]
```

Как мы видим, наибольший ($>10$ ) показатель VIF у **Medu, Fedu, Fjob, famrel, freetime, goout**.

## 14. Сколько признаков достаточно для объяснения 90% дисперсии после применения метода PCA?
***

При выполнении данного задания я ссылаюсь на [эту](https://www.machinelearningmastery.ru/pca-using-python-scikit-learn-e653f8989e60/) статью.

Сначала стандартизируем данные:

```python
from sklearn import preprocessing
stand_X = pd.DataFrame(preprocessing.scale(x_train),columns = x_train.columns)
```

Теперь применим PCA так, чтобы он объяснил 90% дисперсии:

```python
from sklearn.decomposition import PCA
pca = PCA(0.90) # 90 процентов дисперсии
pca.fit(stand_X)
print(pca.n_components_)
print(sum(pca.explained_variance_ratio_))
```

Альтернативно это можно сделать перебором, используя код с лекций:

```python
for i in range(1, len(stand_X.columns)):
    pca = PCA(n_components=i)
    pca.fit(stand_X)
    print(i, sum(pca.explained_variance_ratio_))
```

Видим, что достаточно **18** признаков.

## 15. Какой признак вносит наибольший вклад в первую компоненту?

```python
# нужное нам число признаков
pca = PCA(n_components=25)
x = pca.fit(stand_X)
```

```python
print(x)
```

```python
first_component = pca.components_[0]
greatest_contribution_i = 0
greatest_contribution = first_component[greatest_contribution_i]

for i in range(0, len(first_component)):
    if  abs(first_component[i]) >  abs(greatest_contribution):
        greatest_contribution = first_component[i]
        greatest_contribution_i = i
print(greatest_contribution_i, greatest_contribution)
```

Наибольший по модулю вклад в первую компоненту вносит **31** (если считать с нуля) признак.
